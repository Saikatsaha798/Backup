# Merging

```python
df1 = pd.DataFrame([[1,8,7],[0,6,7],[4,3,1]], columns=['A','B','C'])
df1
```

```python
df2 = pd.DataFrame([[5,8,5],[4,6,1],[2,3,7]], columns=['D','E','F'])
df2
```

```python
pd.merge(df1,df2, right_on='E', left_on='B')
```
