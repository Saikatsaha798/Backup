```python
def binary_search(l, a):
    print(l)
#     print(l[(len(l)//2)-1])
    if len(l) != 1:
        mid = l[(len(l)//2)-1]
        print(mid, a, mid == a)
        if mid == a:
            return True
        elif mid > a:
            binary_search(l[:(len(l)//2)-1], a)
        else:
            binary_search(l[(len(l)//2):], a)
    else:
        return False
```
