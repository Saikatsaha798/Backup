```python
%matplotlib inline                 
from matplotlib import pyplot as plt     #(%matplotlib inline )TO SHOW IN JUPYTER
plt.plot([1,7,8],[7,9,4])
plt.show()
```

```python
from matplotlib import pyplot as plt
plt.plot([1,5,8],[7,6,4])
plt.title("Random Data")     #HEADING
plt.xlabel("X Data")         #X AXIS TITLE
plt.ylabel("Y Data")         #Y AXIS TITLE  
plt.show()
```

```python
from matplotlib import pyplot as plt
plt.plot([1,7,8],[7,9,4], label = 'First', linewidth=2)
plt.plot([1,5,8],[7,6,4], label = 'Second')
plt.title("Random Data")
plt.xlabel("X Data")
plt.ylabel("Y Data")
plt.legend()                #TO SHOW LEGEND
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style as st
st.use("ggplot")         #style
plt.plot([1,7,8],[7,9,4], label = 'First', linewidth=2)
plt.plot([1,5,8],[7,6,4], label = 'Second')
plt.title("Random Data")
plt.xlabel("X Data")
plt.ylabel("Y Data")
plt.legend()                #TO SHOW LEGEND
plt.show()
```

```python
from matplotlib import pyplot as plt
plt.scatter([1,7,8],[7,9,4], label = 'First') 
plt.hist([1,5,8], label = 'Second')
plt.title("Random Data")
plt.xlabel("X Data")
plt.ylabel("Y Data")
plt.legend()                #TO SHOW LEGEND
plt.show()
```
